MainQuest Q10. 입출금 내역
입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.


```python
import random as r

class Account() :

  bank_name = 'SC은행'
  account_count = 0
  # 예금주 : owner / 잔액 : balance
  def __init__(self, owner, balance) :
    self.owner = owner
    self.balance = balance
    self.account_num = self.create_account_num()
    self.deposit_history = []   # 입금내역
    self.withdraw_history = []  # 출금내역
    self.interest_history = []  # 이자내역
    Account.account_count += 1

  # 계좌 생성
  def create_account_num(self) :
    r.seed()
    first_num = r.randint(100, 999)
    mid_num = r.randint(10, 99)
    end_num = r.randint(100000, 999999)
    account_num = f"{first_num}-{mid_num}-{end_num}"
    return account_num

  # 입금
  def deposit(self, money) :
    if money > 0 :
      self.balance += money
      self.deposit_history.append(f"입금 : {self.format_balance_history(money)}원")
      # 입금 내역이 5회 이상일 경우를 5로 나누어 떨어지는 숫자로 구하고 거기에 1% 이자를 더한다.
      if len(self.deposit_history) % 5 == 0 :
        interest = self.balance * 0.01
        self.balance += interest
        self.deposit_history.append(f"이자 : {self.format_balance_history(interest)}원")
        self.interest_history(interest)

  # 출금
  def withdraw(self, money) :
    if money > 0 and self.balance >= money :
      self.balance -= money
      self.withdraw_history.append(f"출금 : {self.format_balance_history(money)}원")

  # 잔고 확인
  def check_balace(self) :
    return self.balance

  # 화면 정보
  def display_info(self) :
    print(f"은행이름 : {self.bank_name}, 예금주 : {self.owner}, 계좌번호 : {self.account_num}, 잔고 : {self.format_balance_num()}원")
    print("거래 내역: ")
    # 입금 내역
    for transaction_details_deposit in self.deposit_history :
      print(transaction_details_deposit)
    # 출금 내역
    for transaction_details_withdraw in self.withdraw_history :
      print(transaction_details_withdraw)
    # 이자 내역
    print("이자 내역: ")
    for transaction_details_interest in self.interest_history :
      print(transaction_details_interest)
    print()

  # 세자리마다 쉼표
  def format_balance_num(self) :
    return '{:,.0f}'.format(self.balance)

  def format_balance_history(self, value)  :
    return '{:,.0f}'.format(value)



# 계좌 추가
accounts = []
accounts.append(Account("이상옥", 5000000))
accounts.append(Account("안중근", 400000))
accounts.append(Account("신채호", 20000))

# 입*출금 test
accounts[0].deposit(50000)
accounts[0].deposit(500000)
accounts[0].withdraw(5000)
accounts[1].deposit(40000)
accounts[1].withdraw(30000)
accounts[2].deposit(10000000)
accounts[2].withdraw(90000)

# 내역 출력
for account in accounts :
  account.display_info()

# 잔고가 100만원 이상인 고객 정보 출력
for account in accounts :
  if account.check_balace() >= 1000000 :
    print("잔고가 100만원 이상인 고객 : ")
    account.display_info()
```

상생 그룹이 +2점을 받게 되면 힌튼 길드와 응길드의 각 길드 스코어는 +2점이 됩니다   

Q11. Aiffel_Group 클래스와 Aiffel_Guild 클래스가 Aiffel클래스의 display_info() 메서드를 아래와 같이 오버라이딩하도록 합니다.   
guild_score(), group2guild_score() 메서드를 호출했을 시 기존 명단에 길드별 스코어 총합이 보이도록 명단을 출력하도록 display_info() 를 오버라이딩합니다.


```python
import random as r

class Aiffel :

  core_dict = {
      "김건": 0, "김경민": 0, "김소연": 0, "김영신": 0, "박근수": 0, "박영수": 0,
      "박영준": 0, "서원영": 0, "신지만": 0, "심지안": 0, "이규상": 0, "이윤상": 0,
      "이진영": 0, "임현석": 0, "전요한": 0, "전재영": 0, "조대희":0
  }

  id_num_count = 1
  quest_socre = {}
  quest_penalty = {}

  def __init__(self, name, ad_year, term, course) :
    self.name = name
    self.ad_year = ad_year
    self.term = term
    self.course = course
    self.id_num = self.generate_id()
    self.quest_socre[self.name] = 0
    self.quest_penalty[self.name] = 0

  # id 생성
  def generate_id(self) :
    last_year_two_num = str(self.ad_year)[-2:]
    term_num = self.term
    course_num = 1 if self.course == "리서치" else 2
    id_num = f"{last_year_two_num}{term_num}{course_num:02d}{Aiffel.id_num_count:02d}"
    Aiffel.id_num_count += 1
    return id_num

  @classmethod
  def assign_id_num(cls, core_dict) :
    for name in core_dict.keys():
      core_dict[name] = cls(name, 2023, 25, "코어").id_num

  # 학번 개수 출력
  @classmethod
  def get_id_num(cls) :
    return cls.id_num_count - 1
  # Aiffel 클래스에 퀘스트 점수 입력을 위한 score 메서드를 추가
  def score(self, score) :
    if 1<= score <= 3 :
      self.quest_socre[self.name] += score
    else :
      print("점수는 1 ~ 3점입니다.")

  def penalty(self, penalty) :
    self.quest_penalty[self.name] += penalty

  def display_info(self) :
    print(f"이름 : {self.name}")
    print(f"학번 : {self.id_num}")
    print(f"과정 종류 : {'리서치' if self.course == 1 else '코어'}")
    print(f"퀘스트 점수 총점 : {self.quest_socre[self.name]}")
    print(f"퀘스트 입력 총점 : {self.quest_socre[self.name] - self.quest_penalty[self.name]}")
    print(f"퀘스트 페널티 총점 : {self.quest_penalty[self.name]}")

  # 그루 정보를 삭제하는 메서드
  @classmethod
  def rm_student(cls, name) :
    del cls.core_dict[name]
    del cls.quest_score[name]
    del cls.quest_penalty[name]


# 랜덤 그룹핑을 해주는 메서드가 추가된 Aiffel_Group 클래스
class Aiffel_Group(Aiffel) :
  group = {
        "상생": [], "신뢰": [], "열정": [], "이끔": [],
        "성장": [], "가치": [], "공유": [], "확산": []
    }

  # 랜덤 그룹핑을 해주는 메서드
  @classmethod
  def random_grouping(cls) :
    core_list = list(cls.core_dict.keys())
    r.shuffle(core_list)
    group_size = len(core_list) // 8
    for i, group_name in enumerate(cls.group.keys()) :
      cls.group[group_name] = core_list[i * group_size: (i + 1) * group_size]

  # Q10. Aiffel_Group 클래스에 group2guild_score 메서드를 추가
  def group2guild_score(cls, gruop_name, score) :
    if gruop_name in cls.group :
      for member in cls.group[gruop_name] :
        for guild_name, guild_member in Aiffel_Guild.guild.items() :
          if member in guild_member :
            Aiffel_Guild.guild_score(guild_name, score)

  # Q11. Aiffel_Group 클래스와 Aiffel_Guild 클래스가
  # Aiffel클래스의 display_info() 메서드를 아래와 같이 오버라이딩
  def display_info(cls) :
    for group_name, group_member in cls.group.items() :
      group_score = 0
      group_penalty = 0
      print(f"{group_name} 그룹 : ")
      for member in group_member :
        group_score += cls.quest_socre[member]
        group_penalty += cls.quest_penalty[member]
      print(f"{group_name} 그룹 총 점수 : {group_score} ({group_score} - {group_penalty})")

# Q8. Aiffel 클래스를 상속하여, 그루 명단을 바탕으로
# “힌튼”, “알트만”, “응”, “하사비스” 네개의 길드로
# 그루들을 랜덤그룹핑해주는 Aiffel_Guild 클래스
class Aiffel_Guild(Aiffel) :
  guilds = {
        "힌튼": [], "알트만": [], "응": [], "하사비스": []
  }

  @classmethod
  def random_guilding(cls) :
    core_list = list(cls.core_dict.keys())
    r.shuffle(core_list)
    guild_size = len(core_list) // 4
    for i, guild_name in enumerate(cls.guilds.keys()) :
      cls.guilds[guild_name] = core_list[i * guild_size: (i + 1) * guild_size]

  # Q9. Aiffel_Guild 클래스에 길드별 점수를 입력하고 차감하는 guild_score 메서드를 추가
  @classmethod
  def guild_score(cls, guild_name, score) :
    if guild_name in cls.guilds :
      for member in cls.guilds[guild_name] :
        cls.quest_socre[member] += score

  @classmethod
  def display_info(cls) :
    for guild_name, guild_member in cls.guilds.items() :
      guild_score = 0
      guild_penalty = 0
      print(f"{guild_name} 길드")
      for member in guild_member :
        guild_score += cls.quest_score[member]
        guild_penalty += cls.quest_penalty[member]
      print(f"{guild_name} 길드 총 점수 : {guild_score} ({guild_score} - {guild_penalty})")


Aiffel.assign_id_num(Aiffel.core_dict)
print(f"학번 개수 : {Aiffel.get_id_num()}")

'''
ex. 모든 길드의 현재 점수가 0점이고 박영수, 박근수, 심지안 그루의 개별 점수가 각각 2, 3, 1 이며
박영수, 심지안 그루는 힌튼 길드, 박근수 그루는 응 길드라고 했을 때, 힌튼길드가 -1점,
응길드가 +1점을 받았다면 길드 점수는 힌튼 -1, 응 +1, 하사비스 0, 알트만 0 이 되어야하고
박영수, 심지안 그루는 각각 1, 0점 박근수 그루는 4점이 되어야 합니다.
'''

student1 = Aiffel("박영수", 2023, 25, "코어")  # 힌튼
student1.score(2)
student2 = Aiffel("박근수", 2023, 25, "코어")  # 응
student2.score(3)
student3 = Aiffel("심지안", 2023, 25, "코어")  # 힌튼
student3.score(1)

student1.display_info()

Aiffel_Group.random_grouping()


Aiffel_Guild.random_guilding()
Aiffel_Guild.display_info()

```
