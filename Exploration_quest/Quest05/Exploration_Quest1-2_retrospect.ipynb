## 3. retrospect

학습이 잘 되어있는 Pre-trained model을 사용하여 재학습을 하지않고, 각각의 모델에 적용하여 성능을 평가하였다.
추가적으로 학습이 잘 되어있는 모델에 lr_schedule이나 early_stopping, dropout을 굳이 새롭게 적용해야 하나 생각이 들었지만 vanilla model만을 적용했을 때 VGG16은 80%의 성능, Resnet50V2는 70%의 성능을 확인하고 새롭게 최적화를 진행하였다.   
그 결과, 정확도는 VGG16가 95%, ResNet50V2는 90%에 도달하였다.   
하지만 새로운 이미지로 테스트를 한 결과, VGG16은 sunflower의 이미지를 제대로 분류하였지만, Resnet50V2는 제대로 분류되지 않는 결과가 나오게 되었다.   
   
이번 프로젝트는 pre-trained model을 가져와서 사용하기에 따로 모델을 설계하는 것보다 성능을 끌어올릴려고 노력하였고, 그 지표들을 보고 어떤 부분에서 튜닝을 진행할지를 고민하였다. 하지만 어떤 수치를 조절해서 성능을 끌어올리는 것 보다 좋은 모델을 선택하고 어떻게  최적화를 하는지에 따라 지표가 눈에 띄게 달라진다는 것을 알게 되었다.
   
이번 프로젝트를 하면서 뭔가 고민하는 부분들이 눈에 보인다라는 느낌을 받게 되었고, 앞으로도 전처리 과정과 모델 설계부분을 자유롭게 조정하는 스킬과 부족한 코딩부분을 보완해야겠다고 느꼈다.
